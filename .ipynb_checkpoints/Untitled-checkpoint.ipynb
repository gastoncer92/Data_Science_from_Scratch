# Finding Key Connectors

Disponemos de una lista de **usuarios**.

```python
users = [
    
    { "id": 0, "name": "Hero" },
    { "id": 1, "name": "Dunn" },
    { "id": 2, "name": "Sue" },
    { "id": 3, "name": "Chi" },
    { "id": 4, "name": "Thor" },
    { "id": 5, "name": "Clive" },
    { "id": 6, "name": "Hicks" },
    { "id": 7, "name": "Devin" },
    { "id": 8, "name": "Kate" },
    { "id": 9, "name": "Klein" }
        ]
```

Y de una lista de relaciones de amistad

```python
friendships = [(0, 1), (0, 2), (1, 2), (1, 3), (2, 3), (3, 4), (4, 5), (5, 6), (5, 7), (6, 8), (7, 8), (8, 9)]
```

Agregamos un campo de amistades a la lista de **usuarios**

```python
for user in users:
    user["friends"]=[]
```

Y luego tomando como modelo $$ friendships = [(i_{1},j_{1}),(i_{2},j_{2})...,(i_{n},j_{n} )] $$

```python
for i,j in friendships:
    users[i]["friends"].append(users[j]["name"]) #en el campo friends del diccionario users, agregar su par ordenado, tomando el nombre 
    users[j]["friends"].append(users[i]["name"]) #se cruzan las ordenes para que los amigos compartan las amistades en común
```

si buscaramos ¿cuál es el número medio de conexiones?, necesitamos el total de conexiones.

```python
def number_of_friends(user,campo):
    return len(user[campo]) # length of friend_ids list
    
total_connections = sum(number_of_friends(user,'friends') for user in users) # 24
```

```python
print(total_connections)
```

```python
num_users = len(users) # longitud de la lista de usuarios
avg_connections = total_connections / num_users # 2.4
```

Esto funciona, pero tiene que examinar toda la lista de intereses para cada búsqueda. Si tenemos muchos
muchos usuarios e intereses (o si queremos hacer muchas búsquedas), probablemente sea mejor
construir un índice desde los intereses hasta los usuarios:

```python
from collections import defaultdict
# keys are interests, values are lists of user_ids with that interest
user_ids_by_interest = defaultdict(list)
for user_id, interest in interests:
    user_ids_by_interest[interest].append(user_id)
```

Y otra de usuarios a intereses:

```python
# keys are user_ids, values are lists of interests for that user_id
interests_by_user_id = defaultdict(list)
for user_id, interest in interests:
    interests_by_user_id[user_id].append(interest)
```

Ahora es fácil encontrar quién tiene más intereses en común con un determinado usuario:
* Iterar sobre los intereses del usuario.
* Para cada interés, iterar sobre los otros usuarios con ese interés.
* Lleva la cuenta de las veces que vemos a cada uno de los otros usuarios.

```python
def most_common_interests_with(user):
    return Counter(interested_user_id
        for interest in interests_by_user_id[user["id"]]
        for interested_user_id in user_ids_by_interest[interest]
        if interested_user_id != user["id"])
```

```python
print(most_common_interests_with(user))
```

# Salaries and Experience

```python
salarios_y_tenencias = [(83000, 8.7), (88000, 8.1),
(48000, 0.7), (76000, 6),
(69000, 6.5), (76000, 7.5),
(60000, 2.5), (83000, 10),
(48000, 1.9), (63000, 4.2)]
```

defaultdict permite al que llama especificar el valor predeterminado por adelantado cuando el contenedor se inicializa.

Por lo general, un diccionario de Python arroja un **KeyError** si intenta obtener un elemento con una clave que no está actualmente en el diccionario. Por el defaultdictcontrario, simplemente creará cualquier elemento al que intente acceder (siempre que, por supuesto, todavía no exista).

```python
# Las claves son los años, los valores son 
# las listas de los sueldos para cada tenencia
salario_por_tenencia = defaultdict(list)  # salario_por_tenencia
for salario, tenencia in salarios_y_tenencias:
    salario_por_tenencia[tenencia].append(salario)
# Las claves son los años, cada valor es 
# el salario medio para esa permanencia
average_salario_por_tenencia = {  # salario_medio_por_tenencia
    tenencia: sum(salarios) / len(salarios)
    for tenencia, salarios in salario_por_tenencia.items()
}
```

Podría ser más útil hacer un cubo con las tenencias:


```python
def tenure_bucket(tenure):
    if tenure < 2:
        return "less than two"
    elif tenure < 5:
        return "between two and five"
    else:
        return "more than five"
```

A continuación, agrupa los salarios correspondientes a cada cubo:

```python
# keys are tenure buckets, values are lists of salaries for that bucket
salary_by_tenure_bucket = defaultdict(list)
for salary, tenure in salarios_y_tenencias:
    bucket = tenure_bucket(tenure)
    salary_by_tenure_bucket[bucket].append(salary)
```

Y, por último, calcula el salario medio de cada grupo:

```python
# keys are tenure buckets, values are average salary for that bucket
average_salary_by_bucket = {
    tenure_bucket : sum(salaries) / len(salaries)
    for tenure_bucket, salaries in salary_by_tenure_bucket.iteritems()
}
```
